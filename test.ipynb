```python
import datetime

class AddressBook:

 person_list = []

#新規にアドレス帳に人を追加する。
 def add(self,person):
    self.person_list.append(person)

#登録された個人一覧を表示する。
 def show_all(self):
    for person in self.person_list:
        print(person.lastname+' '+person.firstname)

#検索条件にマッチする人を表示する。
 def search(self,keyword):
    for person in self.person_list:
     if keyword in person.firstname or keyword in person.lastname:
         print (person.firstname+''+person.lastname)

class Person:
    firstname=''
    lastname=''
    tel=''
    mail_address=''
    birthday=datetime.datetime(2000,1,1)

ab=AddressBook()

p=Person()
p.firstname='はるみ'
p.lastname='いとう'
p.tel='08067431055'

ab.add(p)

p2=Person()
p2.firstname='Ito'
p2.lastname='Harumi'
p2.tel='0456621075'

print('動作確認')
ab.add(p2)

print('アドレス帳の人数->'+str(len(ab.person_list))+'人')

print('一覧表示')
ab.show_all()

print('検索')
ab.search('Harumi')
```
